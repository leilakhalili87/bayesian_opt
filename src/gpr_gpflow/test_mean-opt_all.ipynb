```python
from gpflow.config.__config__ import default_float, default_int
from gpflow.base import Parameter

class Poly1(gpflow.mean_functions.MeanFunction):
    """
    y_i = A x_i + b
    y_i = A x_i^2 + B X_i + c
    """

    def __init__(self, A1=None, B1=None, b1=None):
        """
        A is a matrix which maps each element of X to Y, b is an additive
        constant.

        If X has N rows and D columns, and Y is intended to have Q columns,
        then A must be [D, Q], b must be a vector of length Q.
        """
        gpflow.mean_functions.MeanFunction.__init__(self)
        A1 = np.ones((1, 1), dtype=default_float()) if A1 is None else A1
        B1 = np.ones((1, 1), dtype=default_float()) if B1 is None else B1
        b1 = np.zeros(1, dtype=default_float()) if b1 is None else b1
        
        self.A1 = Parameter(np.atleast_2d(A1))
        self.B1 = Parameter(np.atleast_2d(B1))
        self.b1 = Parameter(b1)


    def __call__(self, X):
        data0 = "../../data/first_cusp.txt"
        data = np.loadtxt(data0)
        ang = data[:,0]
#         ang = (ang - np.mean(ang)) / np.std(ang)
        x_new = X.numpy().astype(int)
        X = ang[x_new].reshape(-1,1)
        X = tf.convert_to_tensor(X, dtype=np.float64)
        var_1 = tf.tensordot(X, self.A1, [[-1], [0]])
        var_d = tf.multiply(tf.math.log(X),X)
        var_2 = tf.tensordot(var_d, self.B1, [[-1], [0]])
        var_3 = self.b1
        return var_1 + var_2 + var_3

class Poly2(gpflow.mean_functions.MeanFunction):
    """
    y_i = A x_i + b
    y_i = A x_i^2 + B X_i + c
    """

    def __init__(self, A2=None, B2=None, b2=None):
        """
        A is a matrix which maps each element of X to Y, b is an additive
        constant.

        If X has N rows and D columns, and Y is intended to have Q columns,
        then A must be [D, Q], b must be a vector of length Q.
        """
        gpflow.mean_functions.MeanFunction.__init__(self)
        A2 = np.ones((1, 1), dtype=default_float()) if A2 is None else A2
        B2 = np.ones((1, 1), dtype=default_float()) if B2 is None else B2
        b2 = np.zeros(1, dtype=default_float()) if b2 is None else b2
        
        self.A2 = Parameter(np.atleast_2d(A2))
        self.B2 = Parameter(np.atleast_2d(B2))
        self.b2 = Parameter(b2)


    def __call__(self, X):
        data0 = "../../data/first_cusp.txt"
        data = np.loadtxt(data0)
        ang = data[:,0].astype(int)
#         ang = (ang - np.mean(ang)) / np.std(ang)
        x_new = X.numpy().astype(int)
        X = ang[x_new].reshape(-1,1)
        X = tf.convert_to_tensor(X, dtype=np.float64)
        var_1 = tf.tensordot(tf.math.subtract(1, X), self.A2, [[-1], [0]])
        var_d = tf.multiply(tf.math.log(X),X)
        var_2 = tf.tensordot(var_d, self.B2, [[-1], [0]])
        var_3 = self.b2
        return var_1 + var_2 + var_3

class Poly3(gpflow.mean_functions.MeanFunction):
    """
    y_i = A x_i + b
    y_i = A x_i^2 + B X_i + c
    """

    def __init__(self, A3=None, B3=None, b3=None):
        """
        A is a matrix which maps each element of X to Y, b is an additive
        constant.

        If X has N rows and D columns, and Y is intended to have Q columns,
        then A must be [D, Q], b must be a vector of length Q.
        """
        gpflow.mean_functions.MeanFunction.__init__(self)
        A3 = np.ones((1, 1), dtype=default_float()) if A3 is None else A3
        B3 = np.ones((1, 1), dtype=default_float()) if B3 is None else B3
        b3 = np.zeros(1, dtype=default_float()) if b3 is None else b3

        self.A3 = Parameter(np.atleast_2d(A3))
        self.B3 = Parameter(np.atleast_2d(B3))
        self.b3 = Parameter(b3)


    def __call__(self, X):
        data0 = "../../data/first_cusp.txt"
        data = np.loadtxt(data0)
        ang = data[:,0].astype(int)
#         ang = (ang - np.mean(ang)) / np.std(ang)
        x_new = X.numpy().astype(int)
        X = ang[x_new].reshape(-1,1)
        X = tf.convert_to_tensor(X, dtype=np.float64)
        var_1 = tf.tensordot(X, self.A3, [[-1], [0]])
        var_d = tf.multiply(tf.math.log(X),X)
        var_2 = tf.tensordot(var_d, self.B3, [[-1], [0]])
        var_3 = self.b3
        return var_1 + var_2 + var_3

class Poly4(gpflow.mean_functions.MeanFunction):
    """
    y_i = A x_i + b
    y_i = A x_i^2 + B X_i + c
    """

    def __init__(self, A4=None, B4=None, b4=None):
        """
        A is a matrix which maps each element of X to Y, b is an additive
        constant.

        If X has N rows and D columns, and Y is intended to have Q columns,
        then A must be [D, Q], b must be a vector of length Q.
        """
        gpflow.mean_functions.MeanFunction.__init__(self)
        A4 = np.ones((1, 1), dtype=default_float()) if A4 is None else A4
        B4 = np.ones((1, 1), dtype=default_float()) if B4 is None else B4
        b4 = np.zeros(1, dtype=default_float()) if b4 is None else b4

        self.A4 = Parameter(np.atleast_2d(A4))
        self.B4 = Parameter(np.atleast_2d(B4))
        self.b4 = Parameter(b4)


    def __call__(self, X):
        data0 = "../../data/first_cusp.txt"
        data = np.loadtxt(data0)
        ang = data[:,0].astype(int)
#         ang = (ang - np.mean(ang)) / np.std(ang)
        x_new = X.numpy().astype(int)
        X = ang[x_new].reshape(-1,1)
        X = tf.convert_to_tensor(X, dtype=np.float64)
        var_1 = tf.tensordot(X, self.A4, [[-1], [0]])
        var_d = tf.multiply(tf.math.log(X),X)
        var_2 = tf.tensordot(var_d, self.B4, [[-1], [0]])
        var_3 = self.b4
        return var_1 + var_2 + var_3

class Poly5(gpflow.mean_functions.MeanFunction):
    """
    y_i = A x_i + b
    y_i = A x_i^2 + B X_i + c
    """

    def __init__(self, A5=None, B5=None, b5=None):
        """
        A is a matrix which maps each element of X to Y, b is an additive
        constant.

        If X has N rows and D columns, and Y is intended to have Q columns,
        then A must be [D, Q], b must be a vector of length Q.
        """
        gpflow.mean_functions.MeanFunction.__init__(self)
        A5 = np.ones((1, 1), dtype=default_float()) if A5 is None else A5
        B5 = np.ones((1, 1), dtype=default_float()) if B5 is None else B5
        b5 = np.zeros(1, dtype=default_float()) if b5 is None else b5

        self.A5 = Parameter(np.atleast_2d(A5))
        self.B5 = Parameter(np.atleast_2d(B5))
        self.b5 = Parameter(b5)


    def __call__(self, X):
        data0 = "../../data/first_cusp.txt"
        data = np.loadtxt(data0)
        ang = data[:,0].astype(int)
#         ang = (ang - np.mean(ang)) / np.std(ang)
        x_new = X.numpy().astype(int)
        X = ang[x_new].reshape(-1,1)
        X = tf.convert_to_tensor(X, dtype=np.float64)
        var_1 = tf.tensordot(X, self.A5, [[-1], [0]])
        var_d = tf.multiply(tf.math.log(X),X)
        var_2 = tf.tensordot(var_d, self.B5, [[-1], [0]])
        var_3 = self.b5
        return var_1 + var_2 + var_3

class Poly6(gpflow.mean_functions.MeanFunction):
    """
    y_i = A x_i + b
    y_i = A x_i^2 + B X_i + c
    """

    def __init__(self, A6=None, B6=None, b6=None):
        """
        A is a matrix which maps each element of X to Y, b is an additive
        constant.

        If X has N rows and D columns, and Y is intended to have Q columns,
        then A must be [D, Q], b must be a vector of length Q.
        """
        gpflow.mean_functions.MeanFunction.__init__(self)
        A6 = np.ones((1, 1), dtype=default_float()) if A6 is None else A6
        B6 = np.ones((1, 1), dtype=default_float()) if B6 is None else B6
        b6 = np.zeros(1, dtype=default_float()) if b6 is None else b6

        self.A6 = Parameter(np.atleast_2d(A6))
        self.B6 = Parameter(np.atleast_2d(B6))
        self.b6 = Parameter(b6)


    def __call__(self, X):
        data0 = "../../data/first_cusp.txt"
        data = np.loadtxt(data0)
        ang = data[:,0].astype(int)
#         ang = (ang - np.mean(ang)) / np.std(ang)
        x_new = X.numpy().astype(int)
        X = ang[x_new].reshape(-1,1)
        X = tf.convert_to_tensor(X, dtype=np.float64)
        var_1 = tf.tensordot(X, self.A6, [[-1], [0]])
        var_d = tf.multiply(tf.math.log(X),X)
        var_2 = tf.tensordot(var_d, self.B6, [[-1], [0]])
        var_3 = self.b6
        return var_1 + var_2 + var_3
```

```python
data_pd = 'metric_euc.txt'
def _octonion_dist(X, X2):
    X = tf.reshape(X, [-1,1])
    pd = np.loadtxt(data_pd)
    X2 = tf.reshape(X2, [-1,1])
    dist0 = np.zeros((len(X), len(X2)))
    dist = tf.Variable(dist0) # Use variable 
    for i in range(len(X)):
        init_val = int(X[i].numpy())
        for j in range(len(X2)):
            fin_val = int(X2[j].numpy())
            dist0[i,j] = pd[init_val, fin_val]
    dist.assign(dist0)
    return dist

def _octonion_dist_single(X):
    X = tf.reshape(X, [-1,1])
    pd = np.loadtxt(data_pd)
    dist0 = np.zeros((len(X)))
    dist = tf.Variable(dist0) # Use variable 
    for i in range(len(X)):
        init_val = int(X[i].numpy())

        dist0[i] = pd[init_val, init_val]
    dist.assign(dist0)
    return dist

class GBKernel(gpflow.kernels.Kernel):
    def __init__(self, variance=1, lengthscales=1, **kwargs):
        super().__init__(**kwargs)
        self.variance = gpflow.Parameter(variance, transform=positive())
        self.lengthscales = gpflow.Parameter(lengthscales, transform=positive())
        self._validate_ard_active_dims(self.lengthscales)
    
    def K(self, X, X2=None):
        X = X[:,0]
        if X2 is None:
            X2 = X
        else:
            X2 = X2[:,0]
        a = self.variance * tf.exp(-_octonion_dist(X, X2)/self.lengthscales)
        tf.debugging.check_numerics(self.lengthscales, 'hi')
        return a

    def K_diag(self, X):
        X = X[:,0]
        a = self.variance * tf.exp(-_octonion_dist_single(X)/self.lengthscales)
        tf.debugging.check_numerics(self.lengthscales, 'hi')
        return a
```

```python
data0 = "../../data/first_cusp.txt"
data = np.loadtxt(data0)
X = data[:, 1:3]
Y = data[:, 3]
Y = (Y - np.mean(Y)) / np.std(Y)
ang = data[:,0].astype(int)


X = X.reshape(-1,2)
Y_all = Y.reshape(-1,1)
# _ = plt.plot(X, Y, "kx", mew=2)


N = X.shape[0]
n_train = 20

# I chose n_train gbs as training data set randomly
# id_train0 = np.sort(np.random.choice(N, size=n_train, replace=False))
id_train0 = np.floor(np.linspace(0,67,20))
id_train1 = np.array([0, 29,43,  67])
id_train0 = np.concatenate((id_train0, id_train1),axis=0)
id_train0 = np.unique(id_train0).astype(int)
id_train = np.hstack([id_train0.reshape(-1,1), X[id_train0][:,1].reshape(-1, 1)])
Y_train = Y_all[id_train[:,0].astype(int)].reshape(-1,1)
# id_train = np.array(id_train, dtype=np.float64)

# # # I chose rest of the gbs as the test set
id_test0 = X[~np.in1d(X[:,0], id_train0)][:,0]
id_test = np.hstack([id_test0.reshape(-1,1), X[id_test0.astype(int)][:,1].reshape(-1, 1)])
Y_test = Y_all[id_test0.astype(int)].reshape(-1,1)

```

```python
k = GBKernel()
# k = gpflow.kernels.Matern12()
pwl = [Poly1(), Poly2(), Poly3(), Poly4(), Poly5(), Poly6()]
meanf = gpflow.mean_functions.SwitchedMeanFunction(meanfunction_list=pwl)
# meanf = None
m = gpflow.models.GPR(data=(id_train, Y_train), kernel=k, mean_function=meanf)
# set_trainable(m.kernel.lengthscales, False)
opt = gpflow.optimizers.Scipy()
opt_logs = opt.minimize(m.training_loss, m.trainable_variables, options=dict(maxiter=100))

mean, var = m.predict_y(id_test)
mean_t, var_t = m.predict_y(id_train)
```
